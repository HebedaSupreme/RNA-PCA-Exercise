```python
def combineAgeSex(x):
    if (x['Age'] == 'Young' and x['Sex'] == "Male"):
        return 0
    if (x['Age'] == "Young" and x['Sex'] == "Female"):
        return 1
    if (x['Age'] == "Aged" and x['Sex'] == "Male"):
        return 2
    if (x['Age'] == "Aged" and x['Sex'] == "Female"):
        return 3
#Young male = 0, Young female = 1, Aged Male = 3, Aged Female = 4
```

```python
dfAnnot['AgeSexCombine'] = dfAnnot.apply(combineAgeSex, axis=1)
#dfAnnot['AgeSexCombine']['RNA 4'] = 3.0
```

```python
from scipy.stats import ttest_ind

cat1A = dfPCA[dfPCA['Age']=='Young']
cat2A = dfPCA[dfPCA['Age']=='Aged']

ttest_ind(cat1A['PC1'], cat2A['PC1'])
```

```python
ttest_ind(cat1A['PC2'], cat2A['PC2'])
```

```python
cat1S = dfPCA[dfPCA['Sex']=='Male']
cat2S = dfPCA[dfPCA['Sex']=='Female']

ttest_ind(cat1S['PC1'], cat2S['PC1'])
```

```python
ttest_ind(cat1S['PC2'], cat2S['PC2'])
```

p value < 0.05 is statistically significant. Null hypothesis is PC1 among Young and PC1 among Old are equal. P-value small means reject the null hypothesis because it is unlikely to be true

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
model1 = ols('PC1 ~ C(Age) + C(Sex)', data=dfPCA).fit()
sm.stats.anova_lm(model1, typ=2)
```

```python
model2 = ols('PC2 ~ C(Age) + C(Sex)', data=dfPCA).fit()
sm.stats.anova_lm(model2, typ=2)
```

```python
model3 = ols('PC1 ~ C(Age) + C(Sex) + C(Age):C(Sex)', data=dfPCA).fit()
sm.stats.anova_lm(model3, typ=2)
```

```python
model4 = ols('PC2 ~ C(Age) + C(Sex) + C(Age):C(Sex)', data=dfPCA).fit()
sm.stats.anova_lm(model4, typ=2)
```

Age is highly significant for PC1 even accounting for Sex. Neither is significant for PC2.

Look at boxplot, trend PC1 is that Age has impact regardless of male or female, but PC2 observes that the pattern between male and female within age groups is opposite. This connection in the PC2 chart suggests the need for further investigation.
